Dropping the unnecessary columns and removing duplicate rows

```python
dfn = dfn.drop('F_CLAIM_HEADER_CLAIM_NUMBER', 1)
dfn = dfn.drop('UNITS_BILLED', 1)
dfn = dfn.drop('LINE_CHARGE', 1)
dfn = dfn.drop('CLAIM_NUMBER', 1)
dfn.drop_duplicates()
dfn
```

```python
df = dfn[dfn['D_DIAGNOSIS'].str.startswith('G35',na=False) | dfn['D_DIAGNOSIS'].str.startswith('C50',na=False) |
         dfn['D_DIAGNOSIS'].str.startswith('H3532',na=False) | dfn['D_DIAGNOSIS'].isnull()]
df = df.drop('DESCRIPTION', 1)
df
```

```python
#Obtaining age of each claimaint in days
df[['DATE_OF_SERVICE','PATIENT_DATE_OF_BIRTH']] = df[['DATE_OF_SERVICE','PATIENT_DATE_OF_BIRTH']].apply(pd.to_datetime) #if conversion required
df['AGE'] = (df['DATE_OF_SERVICE'] - df['PATIENT_DATE_OF_BIRTH']).dt.days
df = df.drop('DATE_OF_SERVICE', 1)
df = df.drop('PATIENT_DATE_OF_BIRTH', 1)
df
```

```python
for col in df:
    print(len(df[col].unique()))
```

```python
for col in df:
    print(df[col].isnull().sum())
```

```python
#Changing the Gender column into two categories 
df['PATIENT_GENDER_CODE'] = df['PATIENT_GENDER_CODE'].map({'1': 'M', '2': 'F','M':'M','F':'F'})
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

Relacing all the categorical nan values with the most frequent value and replacing the numerical nan values with its mean

```python
df.HEALTHGRAPH_ID.fillna(df.HEALTHGRAPH_ID.value_counts().index[0],inplace=True)
df.CLASSIFICATION.fillna(df.CLASSIFICATION.value_counts().index[0],inplace=True)
df.D_PROCEDURE_CODE.fillna(df.D_PROCEDURE_CODE.value_counts().index[0],inplace=True)
df.PROCEDURE_SEQUENCE.fillna(df.PROCEDURE_SEQUENCE.value_counts().index[0],inplace=True)
df.PATIENT_GENDER_CODE.fillna(df.PATIENT_GENDER_CODE.value_counts().index[0],inplace=True)
df.AGE.fillna(df.AGE.value_counts().mean(),inplace=True)
```

```python
for col in df:
    print(df[col].isnull().sum())
```

Label Encoding each categorical column

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df.HEALTHGRAPH_ID = lb.fit_transform(df.HEALTHGRAPH_ID)
df.CLASSIFICATION = lb.fit_transform(df.CLASSIFICATION)
df.D_PROCEDURE_CODE = lb.fit_transform(df.D_PROCEDURE_CODE)
df.PATIENT_GENDER_CODE = lb.fit_transform(df.PATIENT_GENDER_CODE)
df.DIAGNOSIS_SEQUENCE = lb.fit_transform(df.DIAGNOSIS_SEQUENCE)
df
```

```python
pre_df = df[df['D_DIAGNOSIS'].isnull()]
pre_df
```

```python
train_df = df[~df['D_DIAGNOSIS'].isnull()]
```

```python
train_df
```

```python
x_train = train_df.drop('D_DIAGNOSIS',axis=1)
x_train = x_train.drop('D_PATIENT_ID',axis=1)
x_train
```

```python
y_train = train_df[['D_DIAGNOSIS']]
y_train
```

Dividing the Classification tasks into three distinct categories

```python
y_train.loc[y_train['D_DIAGNOSIS'].str.startswith('C50'), 'D_DIAGNOSIS'] = 'C50'
y_train.loc[y_train['D_DIAGNOSIS'].str.startswith('H3532'), 'D_DIAGNOSIS'] = 'H3532'
y_train
```

Label Encoding the Categorical data

```python
y_train.D_DIAGNOSIS = lb.fit_transform(y_train.D_DIAGNOSIS)
y_train
```

'C50' : 0 , 'G35' : 1 , 'H3532' : 2

```python
x_test = pre_df.drop('D_DIAGNOSIS',axis=1)
x_test = x_test.drop('D_PATIENT_ID',axis=1)
x_test.reset_index(drop=True, inplace=True)
x_test
```

Dividing the known diagnosis data into training testing set of 70 percent to 30 percent for training and testing the predictive model

```python
from sklearn.model_selection import  train_test_split
iv=x_train.columns.tolist()
idep=x_train[iv].values
dep = y_train[y_train.columns.tolist()].values
x_tr, x_te, y_tr, y_te = train_test_split(idep,dep, test_size=0.30, random_state=0)
```

```python
x_tr.shape ,y_tr.shape, x_te.shape, y_te.shape
```

Now the feature space becomes a 7 dimensional space with labels being a one dimensional space of multi class classification

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import metrics
```

```python
RF_pipe = make_pipeline(StandardScaler(),RandomForestClassifier())
RF_pipe.fit(x_tr, y_tr)
RF_train_y_pred = RF_pipe.predict(x_te)
print(RF_pipe.score(x_te, y_te))
print('RMSLE:', 100*np.sqrt(metrics.mean_squared_log_error(y_te, RF_train_y_pred)))
```

```python
DT_pipe = make_pipeline(StandardScaler(),DecisionTreeClassifier(random_state=1))
DT_pipe.fit(x_tr, y_tr)
DT_train_y_pred = DT_pipe.predict(x_te)
print(DT_pipe.score(x_te, y_te))
print('RMSLE:', 100*np.sqrt(metrics.mean_squared_log_error(y_te, DT_train_y_pred)))
```

```python
SGD_pipe = make_pipeline(StandardScaler(),SGDClassifier())
SGD_pipe.fit(x_tr, y_tr)
SGD_train_y_pred = SGD_pipe.predict(x_te)
print(SGD_pipe.score(x_te, y_te))
```

Training three different classifiers and testing it RandomForestClassifier is working at maximum accuracy.So to find the Diagnosis RandomForestClassifier can be used.

```python
iv=x_test.columns.tolist()
idep=x_test[iv].values

y_pre = RF_pipe.predict(idep)
```

```python
from collections import Counter
ans={}
for i in range(0,len(pt_id)):
    t = pre_df[(pre_df[['D_PATIENT_ID']] == pt_id[i]).all(1)].index.tolist()
    t1 = y_pre[t]
    data = Counter(t1)
    l = data.most_common(1)[0][0]
    if (l==0):
        ans[pt_id[i]] = 'Breast Cancer'
    elif (l==1):
        ans[pt_id[i]] = 'Multiple Sclerosis'
    elif (l==2):
        ans[pt_id[i]] = 'Wet AMD'
```

Applying Class balance techniques and retraining

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(x_tr, y_tr)

```

```python
RF_pipe = make_pipeline(StandardScaler(),RandomForestClassifier())
RF_pipe.fit(x_smote, y_smote)
RF_train_y_pred = RF_pipe.predict(x_te)
print(RF_pipe.score(x_te, y_te))
print('RMSLE:', 100*np.sqrt(metrics.mean_squared_log_error(y_te, RF_train_y_pred)))
```

```python
metrics.f1_score(y_te,RF_train_y_pred,average='micro')
```
